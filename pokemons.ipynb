Filtrando dados:

```python
#fogo E agua

novo_df = df.loc[(df['Tipo 1'] == 'Fogo') & (df['Tipo 2'] == 'Agua')]
novo_df
```

```python
#fogo OU agua

novo_df = df.loc[(df['Tipo 1'] == 'Fogo') | (df['Tipo 2'] == 'Agua')]
novo_df
```

```python
#filtrando versões MEGA

df.loc[df['Nome'].str.contains('Mega ')]
```

```python
#filtrando versões não mega, só colocar o ~

df.loc[~df['Nome'].str.contains('Mega ')]
```

```python
#expressoes para pokemons de agua e fogo

import re 

df.loc[df['Tipo 1'].str.contains('fogo|agua',flags=re.I,regex=True)]
```

```python
#pokemons que possuem pi

novo_df = df.loc[df['Nome'].str.contains("Pi", flags=re.I, regex=True)]
novo_df
```

```python
#pokemons que começam com pi

novo_df = df.loc[df['Nome'].str.contains("^Pi", flags=re.I, regex=True)]
novo_df
```

```python
#pokemons que terminam com pi

novo_df = df.loc[df['Nome'].str.contains("pi$", flags=re.I, regex=True)]
novo_df
```

Estatisticas agrupadas (GROUPBY)

```python
import pandas as pd

df = pd.read_csv("modificado.csv")
df
```

```python
df = df.drop(columns=['Unnamed: 0'])
numerical_cols = df.select_dtypes(include='number').columns

# Incluindo a coluna de agrupamento
numerical_cols = list(numerical_cols) + ['Tipo 1']

# Aplicando o groupby e calculando a média apenas nas colunas numéricas
df_grouped = df[numerical_cols].groupby(['Tipo 1']).mean().sort_values('Vida', ascending=False)
# Verificando o resultado
df_grouped
```

```python
df['Contador']=1

df.groupby(['Tipo 1', 'Tipo 2']).count()['Contador']
```

Plotar os dados:


```python
#histograma

import matplotlib.pyplot as plt 

df = pd.read_csv("modificado.csv")

atributo = "Ataque"

bins = range(0,201,20)
plt.xticks(bins)

plt.title(f'Distribuição dos Pokemons por {atributo}')
plt.xlabel(f'Valor do {atributo}')
plt.ylabel(f"Quantidade de Pokemons")

plt.hist(df[atributo], bins=bins)
plt.show()
```

```python
atributo = "Total"

bins = range(0,901,100)
plt.xticks(bins)

plt.title(f'Distribuição dos Pokemons por {atributo}')
plt.xlabel(f'Valor do {atributo}')
plt.ylabel(f"Quantidade de Pokemons")

plt.hist(df[atributo], bins=bins)
plt.show()
```

```python
#barra vertical

atributo = "Ataque"

df = df.dropna(subset=['Ataque'])

# Agrupa por 'Tipo 1' e calcula a média para a coluna 'Ataque'
tipos = df.groupby('Tipo 1')[atributo].mean().sort_values(ascending=False)

# Obtém os índices (linhas) e as colunas de interesse
linha = tipos.index.tolist()
coluna = tipos.tolist()

# Cria a figura e o gráfico de barras
figura = plt.figure(figsize=(14, 4))
ax = figura.add_axes([0, 0, 1, 1])
ax.bar(linha, coluna)
plt.xlabel("Tipos de Pokemons")
plt.ylabel(f"Valor médio do atributo {atributo}")
plt.title(f"Valor médio do {atributo} por tipo de Pokemons")

for i,v in enumerate(coluna):
    plt.annotate(f"{v:.2f}", (i-0.3, v+1))
    
plt.show()

```
